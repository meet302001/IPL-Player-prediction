## Handling Categorical Features

```python
cate_list = ["AGE","COUNTRY","PLAYING ROLE","CAPTAINCY EXP"]

for i in cate_list:
    print(df[i].unique())

cate_list_encoded = pd.get_dummies(df[x_feat],columns=cate_list, drop_first = True)
cate_list_encoded
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
data = df[df.iloc[:,4:-3].columns]
scaled_data = pd.DataFrame(scale.fit_transform(data),columns = data.columns)
scaled_data
```

```python
import matplotlib.pyplot as plt

plt_df = scaled_data
fig,ax = plt.subplots()
ax.boxplot(plt_df)
ax.set_title("Checking for Outliers")
ax.set_xlabel("Xlabel")
plt.savefig("mygraph.png")
```

## Train Test Split


```python
#!pip install statsmodels
from sklearn.model_selection import train_test_split
from statsmodels import api as sm

x = cate_list_encoded
y = df["SOLD PRICE"]

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.8,random_state = 12)
```
